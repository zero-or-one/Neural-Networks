```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
from random import uniform
from random import gauss
```

```python
# Helper funcitons

# u(t)
def voltage(V, tau, t):
    e = math.exp(-t/tau)
    return V*(1-e)

# N
def num_per_sec(V, tau, treshold):
    ln = math.log(1 - treshold/V)
    return 1/(-tau*ln)

# draw u(t) versus t for different time range
def u_vs_t(V, tau):
    t = []
    u = []
    time = 0
    while (time <= 5):
        t.append(time)
        u.append(voltage(V, tau, time))
        time += 0.1
    
    plot1 = plt.figure(1)
    plt.ylabel('millivolts')
    plt.xlabel('seconds')
    plt.plot(t, u)
    plt.show()
    
# draw num/time versus t for different time range  
def num_vs_v(treshold, tau, v0):
    V = []
    n = []
    volt = v0
    while (volt <= 2000):
        V.append(volt)
        n.append(num_per_sec(volt, tau, treshold))
        volt += 1
    
    plot1 = plt.figure(2)
    plt.ylabel('number')
    plt.xlabel('millivolts')

    plt.plot(V, n)
    plt.show()
    
def num200_vs_v(tau_mean, tau_std, treshold_mean, treshold_std, v0):
    V = []
    n = []
    volt = v0
    while (volt <= 5000):
        V.append(volt)
        num = 0
        for i in range(200):
            tau = gauss(tau_mean, tau_std)
            treshold = gauss(treshold_mean, treshold_std)
            num =num + num_per_sec(volt, tau, treshold)
        n.append(num)
        volt += 1

    plot1 = plt.figure(3)
    plt.ylabel('number')
    plt.xlabel('millivolts')

    plt.plot(V, n)
    plt.show()
```

```python
# Problem 2(b)

V_1 = 500
tau_1 = 1

V_2 = 50
tau_2 = 0.003

V_3 = 1000
tau_3 = 0.9

V_4 = 2000
tau_4 = 2

V_5 = 1200
tau_5 = 0.2

u_vs_t(V_1, tau_1)
u_vs_t(V_2, tau_2)
u_vs_t(V_3, tau_3)
u_vs_t(V_4, tau_4)
u_vs_t(V_5, tau_5)
```

```python
# Problem 2(d)

treshold_1 = 599
tau_1 = 0.9

treshold_2 = 999
tau_2 = 5

treshold_3 = 1899
tau_3 = 3

treshold_4 = 200
tau_4 = 0.2

treshold_5 = 1500
tau_5 = 0.1

num_vs_v(treshold_1, tau_1, 600)
num_vs_v(treshold_2, tau_2, 1000)
num_vs_v(treshold_3, tau_3, 1900)
num_vs_v(treshold_4, tau_4, 201)
num_vs_v(treshold_5, tau_5, 1501)
```
