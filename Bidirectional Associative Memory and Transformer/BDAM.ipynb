```python
# 06/01/21
# 0|1
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
# helper functions for part a

def gen_binary():
    x = random.rand()
    if x < 0.5:
        return -1
    else:
        return 1

def gen_data(size):
    X = []
    Y = []
    for i in range(size):
        x = []
        y = []
        for j in range(512):
            x.append([gen_binary()])
            y.append([gen_binary()])
            y.append([gen_binary()])
        X.append(x)
        Y.append(y)
    return X, Y
```

```python
# Problem 1 a
x1, y1  = gen_data(50)
x2, y2  = gen_data(100)
x3, y3  = gen_data(200)
```

```python
# Problem 1 b
def transfer(x, y):
    w = 0
    xr = []
    yr = []
    #print(np.shape(x))
    #print(np.shape(y))
    for i in range(len(x)):
       # w += np.matmul(x[i],np.transpose(y[i]))
        w += x[i]*np.transpose(y[i])
    for i in range(len(x)):        
        yi = np.sign(np.matmul(np.transpose(w), x[i]))
        xi = np.sign(np.matmul(w, y[i]))
        xr.append(xi)
        yr.append(yi)
    # print("in fun")
    return xr, yr
```

```python
# helper function for problem c

def error_average(r, r_hat):
    count = []
    for i in range(len(r)):
        c = 0
        for j in range(len(r[0])):
            if (r[i])[j] != (r_hat[i])[j]:
                #print("here")
                c+=1
        count.append(c)
    return np.mean(count)

def BAM(x, y, num):
    xi = x
    yi = y
    avg = []
    for i in range(num):
        xi, yi = transfer(xi, yi)
        avg.append(error_average(yi, y))
    plot2 = plt.figure(1)
    plt.plot(np.arange(1, num+1), avg, linestyle='solid' )
    plt.show()
```

```python
# Problem 1 c
BAM(x1, y1, 10)
```

```python
def poison(x, m):
    xi = []
    arr=random.randint( len(x[0]), size=(m))
    for s in range(len(x)):
        xx = []      
        for i in range(len(x[0])):
            #print(x[s])
            num = x[s][i]
            #print(x[s][i])
            if i in arr:
                #print(num)
                num = np.negative(num)
                xx.append(num)
            else:
                xx.append(num)
        xi.append(xx)
    #print(np.shape(xi))
    return xi 

def noisyBAM(x, y, num):
    total_count = []
    xi = x
    yi = y
    avg = []
    w = np.ones((512, 1024))
    for i in range(num):
        xi, yi= transfer(xi, yi)
        avg.append(error_average(yi, y))
    return avg
```
